```python
# library
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

#**Data Cleaning**

```python
# Checking na values
df_raw.isna().sum()
```

```python
df_raw.isna().sum().sum()
```

```python
# Cleaning data
# Karena rownya sedikit, maka kita akan meng-impute value-value na tersebut dengan nilai mediannya
df_clean1 = df_raw
df_clean1['Distance'].fillna(df_clean1['Distance'].median(), inplace=True)
df_clean1['Bedroom2'].fillna(df_clean1['Bedroom2'].median(), inplace=True)
df_clean1['Bathroom'].fillna(df_clean1['Bathroom'].median(), inplace=True)
df_clean1['Car'].fillna(df_clean1['Car'].median(), inplace=True)
df_clean1['Landsize'].fillna(df_clean1['Landsize'].median(), inplace=True)
df_clean1['BuildingArea'].fillna(df_clean1['BuildingArea'].median(), inplace=True)
df_clean1['YearBuilt'].fillna(df_clean1['YearBuilt'].median(), inplace=True)
df_clean1['Lattitude'].fillna(df_clean1['Lattitude'].median(), inplace=True)
df_clean1['Longtitude'].fillna(df_clean1['Longtitude'].median(), inplace=True)
```

```python
df_clean1.info()
```

```python
# Menurut kami kolom unnamed tidak penting, maka kolom tersebut akan kami drop
df_clean2 = df_clean1.drop(['Unnamed: 0'], axis = 1)
df_clean2.head()
```

```python
df_clean2.info()
```

#**Remove outliers**

```python
# > pilih feature yang ingin di remove outliers
features = ['Rooms', 'Bedroom2', 'Bathroom', 'Car', 'Distance', 'Longtitude', 'Lattitude', 'YearBuilt', 'Price']
df_clean3 = df_clean2[features]
```

```python
# Detect Outliers Using Z-Score
from scipy.stats import zscore
df_clean3['z_scores Rooms'] = zscore(df_clean3['Rooms'])
df_clean3['z_scores Bedroom2'] = zscore(df_clean3['Bedroom2'])
df_clean3['z_scores Bathroom'] = zscore(df_clean3['Bathroom'])
df_clean3['z_scores Car'] = zscore(df_clean3['Car'])
df_clean3['z_scores Distance'] = zscore(df_clean3['Distance'])
df_clean3['z_scores Longtitude'] = zscore(df_clean3['Longtitude'])
df_clean3['z_scores Lattitude'] = zscore(df_clean3['Lattitude'])
df_clean3['z_scores YearBuilt'] = zscore(df_clean3['YearBuilt'])
```

```python
df_clean3.head()
```

```python
# Menghilangkan Outlier pada data dengan menggunakan Z-Score untuk Value Z-score yang kurang dari 3 atau lebih dari 3
condition = (df_clean3['z_scores Rooms'] <= 3) & (df_clean3['z_scores Rooms'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores Bedroom2'] <= 3) & (df_clean3['z_scores Bedroom2'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores Bathroom'] <= 3) & (df_clean3['z_scores Bathroom'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores Car'] <= 3) & (df_clean3['z_scores Car'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores Distance'] <= 3) & (df_clean3['z_scores Distance'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores Longtitude'] <= 3) & (df_clean3['z_scores Longtitude'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores Lattitude'] <= 3) & (df_clean3['z_scores Lattitude'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

```python
condition = (df_clean3['z_scores YearBuilt'] <= 3) & (df_clean3['z_scores YearBuilt'] >= -3)
df_clean3 = df_clean3[condition]
df_clean3 = df_clean3.reset_index(drop=True)
```

#**Building Model**

```python
#splitting dataset
from sklearn.model_selection import train_test_split
```

```python
#kami membagi dataset menjadi data untuk training, validation & testing dengan porsi 70% training dan 30% untuk validation & testing
df_train, df_validation_testing = train_test_split(df_clean4, test_size=0.3, random_state=99)
#kami membagi data untuk validation&testing dengan porsi 60% dan 40% untuk validation dan testing
df_validation, df_testing = train_test_split(df_validation_testing, test_size=0.4, random_state=99)

print("all dataset: ", len(df_clean4))
print("train dataset: ", len(df_train))
print("val dataset: ", len(df_validation))
print("testing dataset: ", len(df_testing))
```

```python
x_train = df_train.drop(['Price'], axis=1) #semua feature di dataset training
y_train = df_train['Price'] #variabel respons untuk dataset training

x_val = df_validation.drop(['Price'], axis=1) #semua feature di dataset validation
y_val = df_validation['Price'] #variabel respons untuk dataset validation

x_testing = df_testing.drop(['Price'], axis=1) #semua feature di dataset testing
y_testing = df_testing['Price'] #variabel respons untuk dataset testing
```

#**Linear Regression**

```python
#build model linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_reg = LinearRegression()

#train
lin_reg.fit(np.array(x_train), y_train)

#evaluation training
y_pred_train = lin_reg.predict(np.array(x_train))
lin_reg_train_mse = mean_squared_error(y_train, y_pred_train)
lin_reg_train_rmse = np.sqrt(lin_reg_train_mse)
lin_reg_r2_train = lin_reg.score(np.array(x_train), y_train)

print("Training RMSE =", lin_reg_train_rmse)
print("Training R^2 =", lin_reg_r2_train)

#evaluation training
y_pred_val = lin_reg.predict(np.array(x_val))
lin_reg_val_mse = mean_squared_error(y_val, y_pred_val)
lin_reg_val_rmse = np.sqrt(lin_reg_val_mse)
lin_reg_r2_val = lin_reg.score(np.array(x_val), y_val)

print("Validation RMSE =", lin_reg_val_rmse)
print("Validation R^2 =", lin_reg_r2_val)
```

#**Decision Tree**

```python
#decision tree

from sklearn.tree import DecisionTreeRegressor
```

```python
#build model Decision Tree Regressor

tree_reg = DecisionTreeRegressor()

#train
tree_reg.fit(np.array(x_train), y_train)

#evaluation training
y_pred_train = tree_reg.predict(np.array(x_train))
tree_reg_train_mse = mean_squared_error(y_train, y_pred_train)
tree_reg_train_rmse = np.sqrt(tree_reg_train_mse)
tree_reg_r2_train = tree_reg.score(np.array(x_train), y_train)

print("training RMSE =", tree_reg_train_rmse)
print("training R^2 =", tree_reg_r2_train)

#evaluation training
y_pred_val = tree_reg.predict(np.array(x_val))
tree_reg_val_mse = mean_squared_error(y_val, y_pred_val)
tree_reg_val_rmse = np.sqrt(tree_reg_val_mse)
tree_reg_r2_val = tree_reg.score(np.array(x_val), y_val)

print("validation RMSE =", tree_reg_val_rmse)
print("validation R^2 =", tree_reg_r2_val)
```
